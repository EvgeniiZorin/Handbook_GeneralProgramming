# Socket

A socket is like a virtual plug that allows two programs to talk to each other - either on the same computer or over a network. 

Types of sockets:
- **TCP (Transmission Control Protocol) Socket**
  - Used for communication over the internet or a network
  - Reliable: ensures that data arrives in order and without errors
  - Example: when you load a website, your browser uses TCP sockets to talk to the server.
- **UDP socket**
- **UNIX Socket**
  - Used for communication between programs on the same computer
  - Faster than TCP because it doesn't go through the network stack
  - Example: a web server like Nginx talking to a backend app like a Python service on the same machine


## Example 1

```python
"""
Put this to the file server.py
"""
import socket

# define host
# method 1
HOST = socket.gethostbyname(socket.gethostname())
print(HOST)
# method 2
# HOST = '192.168.100.15' # or 'localhost'
HOST = 'localhost'
# define port
PORT = 34

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind((HOST, PORT))
# listen to incoming connections
server.listen(5) # allow up to 5 connections

while True:
    communication_socket, address = server.accept()
    print(f"Connected to {address}")
    message = communication_socket.recv(1024).decode('utf-8') # 1024 bytes
    print(f"Message from client is: {message}")
    communication_socket.send(f"Got your message, thank you!".encode('utf-8'))
    communication_socket.close()
    print(f"Connection with {address} ended")
```

```python
"""
Put this to the file client.py
"""
import socket

HOST = '192.168.100.15'
PORT = 34

socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
socket.connect((HOST, PORT))

socket.send('Hello world').encode('utf-8')
print(socket.recv(1024))
# print(socket.recv(1024).decode('utf-8'))
```
